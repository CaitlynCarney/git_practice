## Relational Plots

```python
# Explore relationship between total_bill and tip

sns.relplot(data = tips, x = 'total_bill', y = 'tip')
# relplot indicated relational plot
# the data we are looking at are tips
# x and y are what we want to plot
# returns a FacetGrid (which means a figure level object)
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip')
plt.title("Total Bill vs. Tip Amount")
#This is just adding a title to the above plot
# does not change anything else about the plot
```

```python
#plt.style.use() was customizationn in matplotlib
#in seaborn we use sns.set()
# used for fonts, colors etc.
sns.set()
sns.relplot(data = tips, x = 'total_bill', y = 'tip')
plt.title("Total Bill vs. Tip Amount")
```

```python
# see distribution of total and tip
# brerak it down by smoker
# to do that we need to visualize if the distribution is different for smoker vs. non smoker
    # do this by adding hue = 'column_name'
sns.set()
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'smoker')
plt.title("Total Bill vs. Tip Amount")
```

```python
#instead of by smoker specify time of day
sns.set()
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day')
plt.title("Total Bill vs. Tip Amount")
#be careful if you have too many catagories it looks messy and unreadable
#try to use hue if you have 2 max 3 different values not more
```

```python
#instead of by smoker or day specify gender
sns.set()
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex')
plt.title("Total Bill vs. Tip Amount")
```

```python
sns.set()
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'time')
plt.title("Total Bill vs. Tip Amount")
```

```python
# there is another key word you can pass
    # style = 'cokumn_name' instead of hue
#style as a k word
sns.set()
sns.relplot(data = tips, x = 'total_bill', y = 'tip', style = 'time')
plt.title("Total Bill vs. Tip Amount")
```

```python
# we can use an arguement called col
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'day')
#will return 4 different subplots broken up by day of the week it is
# give you similar figure with different subl=plotts in it
```

```python
#use col_wrap to give you same subplot in specified amount of columns
    # col_wrap = number
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'day', col_wrap = 2)
#use col_waro to specify columns
```

```python
# can stiol use hue with col_wrap
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'day',hue = 'time', col_wrap = 2)
```

### Key takeaways?

- tip amount generally increases with total_bill amount
- the waiter reporting didnt work lunch on Saturday or Sunday, 
- the waiter works during lunch on Thursday
- the waiter works both lunch and dinner time on Friday

### Regression plots

```python
# Add Regression line to relplot by using sns.lmplot

sns.lmplot(x = 'total_bill', y = 'tip', data = tips)

#Translucant band is 95% CI and you can change this key word arguemtn to change translucently
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', col = 'smoker', data = tips)
#col = 'column_name' will add 2 subplots one for smoker and another for non smoker
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips)
#hue = 'column_name' will add 2 lines 1 for smoker and another for nonsmoker
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'sex', data = tips)
```

### Categorial Plots


**Boxplots**

```python
#sns.boxplot() will create the boxplot
# vertical boxplot
sns.boxplot(data = tips, y = 'tip')
```

```python
# horizontal boxplot
# just change y to x
sns.boxplot(data = tips, x = 'tip')
# the points outside the boxplot are outliers
```

####                 Boxplot Diagram Labled For Extra Information

```python
# adding ticks and a grid to prior boxplot to make it more clear
sns.boxplot(data = tips, x = 'tip')
plt.grid(True, ls = '--')
plt.xticks(np.arange(0,10,1));
```

```python
tips.tip.describe()
#gives you all the values so you dont have to eyeball it
```

```python
#make different kind of subplots

plt.figure(figsize = (10,8))

plt.subplot(221)
#plt.subplot(2colums2rows1stsubplot)
sns.boxplot(y = 'tip', data = tips)
#this one will be vertical

plt.subplot(222)
sns.boxplot(x = 'tip', data = tips)
#this one will be hrizontal

plt.subplot(223)
sns.boxplot(y = 'tip', x = 'size', data = tips)
#will also break it down by party size on top of everythin else

plt.subplot(224)
#to change the order for dinner vs lunch opass and argument called order afolowed by a list of how you want it ordered
sns.boxplot(x = 'tip', data = tips, y = 'time', hue = 'smoker', order = ['Lunch', 'Dinner'])
#will have distribution broken down by dinner and lunch
#hue will also break it down further into smoker or non smoker
# hue can also chagne the order by doing hue_order = ['Yes', 'No']
```

**Heatmap**

```python
data = pd.crosstab(tips.time, tips['size'])
```

```python
# to create a heatmp use sns.heatmap()
sns.heatmap(data, cmap = plt.cm.Greens, annot = True, fmt = 'd')
```

```python
data = pd.crosstab(tips.time, tips.smoker)
data
```

```python
sns.heatmap(data, annot = True, fmt = 'd')
```

**Pairplot**

```python
#one of the most useeful ones
#give a broad veiw
#can quickly look up correlations with this one
# simple to use
```

```python
# to create o a pair plot use sns.pairplot(column_name)
sns.pairplot(tips)
# takes in all variables and gives this different plot forms
#useful to just eyebacll and see relationships between graphs
```

```python
sns.pairplot(tips, corner = True)
# corner = true just removes duplicates
```

```python
#break it down by hue
sns.pairplot(tips, hue = 'smoker')
#now it is broken down by smoker vs non msoker
```

- very useful in machine learning
- in some cases you may havee a ot of variables like 30 or more making a pair map may be messy so you may want to cr4eate a new data set only containing the varaibles you wan to look at then pass that dataset onto to the pairplot

**Jointpplot**

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
# if you want to look at the  regression line pass on kind = 'reg'
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'reg')
# it now has the regression and kd line
```

```python

sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hex')
# divides chart area into hexagonal figures and colors them in
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde')
#gives you back a 2-dimensional visual
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde', fill = True)
#gives you back a 2-dimensional visual filled in
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde', cmap = "magma")
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde', cmap = "magma", fill = True)
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde', cmap = "magma", fill = True, thresh = 0)
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde', cmap = "magma", fill = True, thresh = 0, levels = 60)
#gives you back a 2-dimensional visual
```
