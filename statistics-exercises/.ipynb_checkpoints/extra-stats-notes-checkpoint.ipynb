# Hypothesis Testing


- **Null Hypothesis (H0):**
    - There is no difference between smokers' tips and the overall population's tip average.


- **Alternative Hypothesis: (ùêªùëé)**
    - There is a difference between smokers' tips and the overall population's tip average.


- **Confidence level:**
    - How confident are you in our results. (90% = .9, 95% =  .95, 99% = .99)


- **Alpha(ùëé):**
    - 1 - confidence level
        - 1 - .9 = .1
            - 10% chance that you will make a Type I Error (False Positive) or reject a True Null hypothesis.
        - 1 - .95 = .05
            - 5% chance that you will make a Type I Error (False Positive) or reject a True Null hypothesis.
        - 1 - .99 = .01
            - 1% chance that you will make a Type I Error (False Positive) or reject a True Null hypothesis.


- **T-statistic:**
    - the calculated difference represented in units of standard error. 
        - The greater the magnitude of t, the greater the evidence against the null hypothesis


- **P-value:**
    - values we obtain from hypothesis testing
        - represents the probability that our obtained result is due to chance given that our stated hypothesis is true.


# Hypothesis Testing Errors

- **Type 1 error:**
    - False Positive
        - Think of a sounding a false alarm.
            - reject the Null hypothesis when the Null hypothesis is actually True.


- **Type 2 error:**
    - False Negative
        - miss identifying a real difference or relationship that exists in reality.
            - I fail to reject the Null hypothesis when the Null hypothesis is actually false.

# T-test - Continuous v Categorical Variables

- **One Sample T-test**
    - is when I compare the mean for a subgroup to the population mean.
        - Are sales for group A higher when we run a promotion?
    - compares the mean of a subgroup with the population mean


`subpop = array, list, Series`

`popmean = single_value`

`t, p = scipy.stats.ttest_1samp(subpop, popmean)`

- **Two Sample T-test**
    - when we compare the mean of one subgroup to the mean of another subgroup.
    - compares the means of two subgroups.
    
`subpop_a = array, list, or Series`

`subpop_b = array, list, or Series`

`t, p = scipy.stats.ttest_ind(subpop_a, subpop_b)`

## Types of t-tests

- **one-tailed test**
    - looks for a specific difference: appropriate if I only want to determine if there is a difference between groups in a specific direction, positive only or negative only.

- Check one-tailed test for significance in positive direction. (greater than)
    - **(p/2) < a and t > 0**

- Check one-tailed test for significance in negative direction. (less than)
    - **(p/2) < a and t < 0**

- **two-tailed test** 
    - looks for any difference: appropriate if I want to test for significance without concern for a positive or negative direction.

- Check a two-tailed test for significance, non-directional.
    - **p < alpha**

## Eample of a 1 sample test

```python
df = data('tips')
df.head()
```

```python
# T-tests assume that the continous variable is normally distributed; quick check of this.

sns.distplot(df.tip)
plt.title('Distribution of Tips - Slight Right Skew')
plt.show()
```

```python
#The distribution is slightly right-skewed, so I'll do a quick comparison of the mean and median of df.tip.
#I can see below that the mean is slightly higher than the median. It's close enough to proceed, though.
print(f'The mean is: {df.tip.mean()}, and the median is {df.tip.median()}')
print('This is close enough to normal to continue.')
```

```python
#Set hypothesis
null_hypo = "There is no difference between smokers' tips and the overall population's tip average."
alt_hypo = "There is a difference between smokers' tips and the overall population's tip average."
#Sell the alph
confidence_interval = .95
alpha = 1 - confidence_interval
print(f'alpha = {alpha}')
```

```python
# Create smokers subset of our df.
smokers = df[df.smoker == 'Yes']

# Assign the mean of all tips to the variable pop_mean.
pop_mean = df.tip.mean()

# Pass in the tip column from smokers subset and mean of tip column from entire df.
t, p = stats.ttest_1samp(smokers.tip, pop_mean)
print(f't = {t:.3f}')
print(f'p = {p:.3f}')
```

```python
print(f'Our p-value is less than our alpha: {p < alpha}')
```
