6. Is the sepal length significantly different in virginica compared to versicolor? 


```python
import scipy.stats as stats
```

    a. Run a statistical experiment to test this.
    b. Make sure to include a null hypothesis, alternative hypothesis, results, and summary.
    c. What is your takeaway from this statistical testing?

```python
null_hypothesis = "there is no difference between sepal_length in virginica and sepal_length in versicolor"
alt_hypothesis = "sepal_length significantly different between virginica and versicolor"
confidence_level = .95
alpha = 1 - confidence_level
```

```python
virginica = df[df.species_name == 'virginica']
versicolor = df[df.species_name == 'versicolor']
```

```python
virginica.sepal_length.var(), versicolor.sepal_length.var()
```

```python
t, p = stats.ttest_ind(virginica.sepal_length, versicolor.sepal_length)
t, p
```

```python
if (p/2) < alpha and t > 0:
    print("We reject the null hypothesis")
    print("We move forward with the understanding that there is a difference in sepal_length")
else:
    print("we fail to reject the null hypothesis")
    print("we dont have sufficent evidence to support the claim that", alternative_hypothesis)
```
