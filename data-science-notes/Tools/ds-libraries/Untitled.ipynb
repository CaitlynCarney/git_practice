# Line Chart
- For the most basic example, we can create a sequence of numbers, then a chart (calling .plot), then call the .show method to display it.

    - Note that the act of creating the chart and showing it are two separate acts.

```python
x = list(range(150))

plt.plot(x)
plt.show()
```

```python
# When only given one variable, it will be treated as y, and a range will be generated for us. 
# We can also explicitly specify the x and y values. We could write the above example like so:

plt.plot(range(len(x)), x)
plt.show()
```

# We can display two series on the same chart by calling .plot multiple times before we call .show.

```python
from random import randint

# Some random data
x1 = [randint(-5, 25) + n for n in range(150)]
x2 = [x + randint(-6, 6) for x in x1]

# we can use the alpha kwarg to define how transparent the color is
plt.plot(x1, c='green', alpha=0.6)
plt.plot(x2, c='red', alpha=0.4)
plt.show()
```

        - Until we call .show, we'll be referring to the same chart, so we can keep adding to it.

# -----------------------Custom Ticks--------------------------------

- We can specify custom ticks with the .xticks and .yticks methods. Both methods require 2 arguments, a list of the indices that the custom ticks will be placed at, and a list of custom ticks.

```python
x = [1, 2, 3.14, 4] # your ticks will show up on 1, 2, 3.14, and 4 on your x axis
y = [n ** 2 for n in x] # your ticks will show up on n^2 (n being x) on your y axis
plt.scatter(x, y)
plt.xticks([1, 2, 3.14, 4], ['one', 'two', '$\pi$', 'four']) #renames x axis lables to one, two, pi, four
plt.show()
```

# --------------------Custome Line Types-------------------------

- We can change the type of the line on a line chart with the ls (line style) keyword argument.

    - ':' represents a dotted line
    - '-' represents a dashed line

```python
import math

x = range(-4, 4) #makes x axis range from -4 to 4
y = [math.sin(n) for n in x] #adjusts y axis range
plt.scatter(x, y)
plt.plot(x, y, ls=':')#makes the line a dotted line
plt.title('An approximation of the sine function')#names it
plt.xlabel('x')#lables x axis
plt.ylabel('sin(x)')#lables y axis
plt.show()
```

# ----------------Rotating Text and Annotations---------------------

- Axis ticks can be rotated with the rotation keyword argument.

```python
x = range(1, 5)#sets x axis range
y = x#sets y axis range
plt.scatter(x, y)
plt.yticks([1, 2, 3, 4], ['one', 'two', 'three', 'four'], rotation=45)#rotation = 45 will rotate lables 45 degrees
plt.xticks(rotation=45)#rotation = 45 will rotate lables 45 degrees
plt.show()
```

- This can be particularly useful for dates, or visualizing time data.

```python
import random

weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'] * 2

x = range(14) #sets x axis range
y = [random.randint(1, 10) for _ in x] #sets y axis range
plt.plot(x, y)
plt.xticks(x, weekdays, rotation=60)#rotates x axis lable 60 degrees
plt.show()
```

# --------------------------Adding Text--------------------------------

- We can use the .text to add arbitrary text to the chart.
    - We'll need to specify x and y coordinates for where the text should be placed. The coordinates provided specify where the left boundary of the text starts.
    - We can also specify the color and fontsize keyword arguments to change the presentation of the text.

```python
x = list(range(-4, 4)) #sets x axis range
x.append(3.14159)
y = [math.sin(n) for n in x] #sets y axis range

plt.scatter(x, y)
plt.plot(x, y, ls=':')

plt.text(0.25, 0, '(0, 0)', fontsize=10, color='blue')
#(0.25, 0, '(0, 0)' sets texts coordinates
#fontsize=10 sets text size
#color='blue' sets text color
plt.text(1.2, 0.1, '($\pi$, sin($\pi$))', fontsize=14, color='firebrick')
#1.2, 0.1, '($\pi$, sin($\pi$))' sets texts coordinates
#fontsize=14 #sets text size
#color='firebrick' sets text color
plt.show()
```

- Annotations can be even more fine-grained with the .annotate method.

```python
x = range(-3, 10) #sets x axis range
y = [n * 2 for n in x] #sets y axis range
z = [n / 2 for n in x]

plt.plot(x, y, c='orange') #sets line color
plt.plot(x, z, c='green') #sets line color

plt.annotate('Intersection', xy=(0, 0), xytext=(-3, 5),
             arrowprops={'facecolor': 'blue'}) #This produces an arrow that points to the 
                                                #point specified with the xy keyword 
                                                #argument, and starts from where the xytext 
                                                #is specified. This arrow will blue and
                                                #labled intersection
plt.show()
```

# -------------------Figure Size and Legends-------------------------

- We can change properties of the figure by accessing the plt.figure method. Keep in mind that we need to specify this before adding to the chart.

- We can add a legend to our chart by specifying labels for the series we are plotting then calling the .legend method.

```python
x = range(-4, 5)#sets x axis range
sin_x = [math.sin(n) for n in x]#create the line plot for sin
cos_x = [math.cos(n) for n in x]#create the line plot for cos

plt.figure(figsize=(14, 6)) #changes the size of plot 
            #figsize=(14, 6) .... setsthe=(width, height)

plt.scatter(x, sin_x, c='blue') #colors the line plot for sin
plt.scatter(x, cos_x, c='orange') #colors the line plot for cos
plt.plot(x, sin_x, c='blue', label='$\sin(x)$') #creates lable for the legend names sin(x) colored blue
plt.plot(x, cos_x, c='orange', label='$\cos(x)$')#creates lable for the legend names cos(x) colored orange

plt.legend(loc='upper right') #sets lengend location
plt.title('Approximate sin and cosine curves') #sets the title
plt.xlabel('$x$') #sets x axis lable

plt.show()
```

# ----------------------------Subplots----------------------------------

- We can create multiple charts together using the .subplot function. We'll need to specify the total number of rows, total number of columns, and which chart is currently the "active" chart.

```python
n_rows = 1 #sets number of rows
n_cols = 2 #sets number of columns

# some data to play with
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 2, 3, 4, 5]

# plot the first subplot
plt.subplot(n_rows, n_cols, 1) #the 1 sets this as the first subplot
plt.plot(x, y) #makes it use the data names x and y
plt.title('x ~ y') #names the subplot

# the second subplot
plt.subplot(n_rows, n_cols, 2) #the 2 sets this as the first subplot
plt.plot(x, z) #makes it use the data names x and z
plt.title('x ~ z') #names the subplot

plt.show()
```

- Using subplots is very common, and the number of rows, columns, and active chart can be be specified in a short hand of just one number.

```python
# some data to play with
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 2, 3, 4, 5]

# plot the first subplot
plt.subplot(121)#first one
plt.plot(x, y)#makes it use the data names x and y
plt.title('x ~ y')#names the subplot

# the second subplot
plt.subplot(122)#second one
plt.plot(x, z)#makes it use the data names x and z
plt.title('x ~ z')#names the subplot

plt.show()
```

- Notice that while we can specify titles for the individual subplots, we can also specify an overall title for the chart with the .suptitle function.

```python
# some data to play with
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 2, 3, 4, 5]

# plot the first subplot
plt.subplot(121)#first one
plt.plot(x, y)#makes it use the data names x and y
plt.title('x ~ y')#names the subplot

# the second subplot
plt.subplot(122)#second one
plt.plot(x, z)#makes it use the data names x and z
plt.title('x ~ z')#names the subplot

plt.suptitle('Subplots Demo') #names all subplots together

plt.show()
```

# --------------------------Histograms---------------------------------

- We can create histograms with the .hist function.

```python
x = [1, 1, 2, 3, 3, 3, 4, 4, 5]

plt.hist(x)
plt.show()
```

If we don't tell matplotlib how to bin the data, it will do it's best to guess for us, but we can also explicitly specify the bins.

```python
x = [1, 1, 2, 3, 3, 3, 4, 4, 5]

plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6]) #lables the bins
plt.show()
```

The align property tells matplotlib how the bins should be aligned over the data that they represent. Here's an example of all 3 possible alignments: left, middle, and right:

```python
x = [1, 1, 2, 3, 3, 3, 4, 4, 5]

plt.figure(figsize=(16, 6))
plt.suptitle(f'x = {x}')

plt.subplot(131)#Makes it first
plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6])
plt.title('align="middle" (default)') #names it

plt.subplot(132)#makes it second
plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6], align='left')
plt.title('align="left"')#names it

plt.subplot(133)#makes it third
plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6], align='right')
plt.title('align="right"')#names it

plt.show()
```

Like scatter and line charts, we can plot multiple histograms together. When doing so, sometimes it is useful to set the alpha, or transparency, for the colors.

We'll show two examples here, one using the same color for each series, one using different colors:

```python
x1 = [randint(1, 5) for _ in range(20)]
x2 = [randint(1, 5) for _ in range(20)]

plt.hist(x1, bins=[0, 1, 2, 3, 4, 5, 6], align='left', edgecolor='black', alpha=0.5, color='blue')
plt.hist(x2, bins=[0, 1, 2, 3, 4, 5, 6], align='left', edgecolor='black', alpha=0.5, color='blue')
#edgecolor sets the edges of the bars
#color sets the fill of the bar
plt.show()
```

```python
x1 = [randint(1, 5) for _ in range(20)]
x2 = [randint(1, 5) for _ in range(20)]

plt.hist(x1, bins=[1, 2, 3, 4, 5, 6], align='left', edgecolor='black', alpha=0.4, label='$x_1$')
plt.hist(x2, bins=[1, 2, 3, 4, 5, 6], align='left', edgecolor='black', alpha=0.4, label='$x_2$')
#alpha is a way of changing color
plt.legend()

plt.show()
```

As an alternative to having the series overlap, we can also plot the series side by side by passing the data series as a tuple to the hist function.

```python
x1 = [randint(1, 5) for _ in range(20)]
x2 = [randint(1, 5) for _ in range(20)]

plt.hist((x1, x2), bins=[1, 2, 3, 4, 5, 6], align='left', edgecolor='black', alpha=0.6, label=['$x_1$', '$x_2$'])

plt.legend()
plt.show()
```

Let's take a look at one more example:

```python
plt.figure(figsize=(12, 8))
for i in range(1, 5):
    n = 10 ** i
    x1 = [randint(1, 6) for _ in range(n)]
    x2 = [randint(1, 6) for _ in range(n)]

    total = [x + y for x, y in zip(x1, x2)]

    plt.subplot(2, 2, i)
    plt.hist(total, bins=range(2, 14), align='left', color='white', edgecolor='black')
    plt.title(f'n = {n}')

plt.suptitle('Sum of 2 dice rolls')
plt.show()
```
